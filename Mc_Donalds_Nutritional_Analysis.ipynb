Analysis

```python
from scipy.stats import zscore
```

```python
menu_fin=menu.drop(["Item","Category","Serving Size"],axis=1)
```

```python
menu_fin1 = menu_fin.apply(zscore)
menu_fin1
```

```python
cov_mat = np.cov(menu_fin1.T)
eig_vals,eig_vecs = np.linalg.eig(cov_mat)
tot =sum(eig_vals)
var = [(i/tot)*100 for i in sorted(eig_vals,reverse=True)]
cumsumvar = np.cumsum(var)
```

```python
cumsumvar
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
menu_pca = pca.fit_transform(menu_fin1)
```

```python
menu_pca1 = pd.DataFrame(menu_pca)
menu_pca1.head(2)
```

```python
menu_pca2=menu_pca1
menu_pca2['Category']=menu['Category']
menu_pca2.head(2)
```

```python
for col in menu_pca2.drop(labels=['Category'], axis=1):
  fig = px.bar(menu_pca2,x='Category', y=col,
             color='Category', height=400)
  fig.show()
```

```python
wsc=[]
from sklearn.cluster import KMeans
for i in range(1,16):
    kmeans= KMeans(n_clusters=i)
    kmeans.fit(menu_pca1)
    wsc.append(kmeans.inertia_)


plt.plot(range(1,16),wsc)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.metrics import silhouette_samples,silhouette_score
kmeans4= KMeans(n_clusters=4)
kmeans4.fit(menu_pca1)
kmeans4.labels_
silhouette_score(menu_pca1,kmeans4.labels_)
```

```python
from yellowbrick.cluster import KElbowVisualizer

model = KMeans()
visualizer = KElbowVisualizer(
    model, k=(2,12), metric='calinski_harabasz', timings=False)

visualizer.fit(menu_pca1)
visualizer.show()
```

```python
from yellowbrick.cluster import SilhouetteVisualizer

model = KMeans()
visualizer = KElbowVisualizer(
    model, k=(2,12), metric='silhouette', timings=False)

visualizer.fit(menu_pca1)
visualizer.show()
```

```python
model = KMeans(4, random_state=42)
visualizer = SilhouetteVisualizer(model, colors='yellowbrick')

visualizer.fit(menu_pca1)
visualizer.show()
```

```python
from yellowbrick.cluster import InterclusterDistance



model = KMeans(3)
visualizer = InterclusterDistance(model)

visualizer.fit(menu_pca1)
visualizer.show()
```

```python
menu_fin["cluster"]= kmeans4.labels_
```

```python
menu["cluster"]=kmeans4.labels_
menu["cluster"].value_counts()
```

```python
menu.groupby("cluster").mean().T
```

```python
menu_pca2['Cluster']=menu['cluster']
menu_pca2.head(2)
```

```python
x_pca=menu_pca2[[0,1]]
y_pca=menu_pca2['Cluster']
```

```python
from imblearn.over_sampling import ADASYN
X, Y = ADASYN().fit_resample(x_pca, y_pca)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.3, random_state=0)
```

```python
y_test.head()
```

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(x_train,y_train)
pred_lr=classifier.predict(x_test)
gauss=classifier.score(x_test, y_test)
gauss
```

```python
print(classification_report(y_test, pred_lr))
```

```python
visualizer = ClassificationReport(classifier,support=True)

visualizer.fit(x_train, y_train)
visualizer.score(x_test, y_test)
visualizer.show()
```

```python
from sklearn import svm
clf = svm.SVC(kernel='linear', C=1.0)
clf.fit(x_train, y_train)
y_pred=clf.predict(x_test)
vec=clf.score(x_test, y_test)
vec
```

```python
print(classification_report(y_test, y_pred))
```

```python

from yellowbrick.classifier import ClassificationReport
visualizer = ClassificationReport(clf,support=True)

visualizer.fit(x_train, y_train)
visualizer.score(x_test, y_test)
visualizer.show()
```

```python
from xgboost import XGBClassifier
model = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')
model.fit(x_train, y_train)
prediction = model.predict(x_test)
xgb=model.score(x_test, y_test)
xgb
```

```python
print(classification_report(y_test, prediction))
```

```python
visualizer = ClassificationReport(model,support=True)

visualizer.fit(x_train, y_train)
visualizer.score(x_test, y_test)
visualizer.show()
```

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import SVC
model = OneVsRestClassifier(SVC())
model.fit(x_train, y_train)
prediction = model.predict(x_test)
one=model.score(x_test, y_test)
one
```

```python
print(classification_report(y_test, prediction))
```

```python
visualizer = ClassificationReport(model,support=True)

visualizer.fit(x_train, y_train)
visualizer.score(x_test, y_test)
visualizer.show()
```

```python
models = pd.DataFrame({'Model': ['gaussian','SVM', 'XGBoost',
                                 'one vs rest'],
                       'Score': [gauss, vec, xgb,one]})

model=models.sort_values(by = 'Score', ascending = False, ignore_index = True)
model
```
